```python
from pathlib import Path

import pandas as pd
```

```python
pattern_p          = Path("../../../data/out/pattern.csv")
out_pattern_p      = Path("../../../data/out/pattern_stats.csv")
rest_pattern_p     = Path("../../../data/out/rest_detail_pattern.csv")
out_rest_pattern_p = Path("../../../data/out/rest_detail_pattern_stats.csv")
```

```python
df = pd.read_csv(pattern_p)
rest_df = pd.read_csv(rest_pattern_p)
```

```python
df_counts = df.value_counts(subset="pattern")
df_counts
```

```python
df_stats = pd.DataFrame([df_counts.index, df_counts.tolist(), (df_counts.tolist())]).T
df_stats.rename(columns={0: "pattern", 1: "count", 2: "frequency"}, inplace=True)
df_stats.frequency = df_stats.frequency /  df.shape[0]
df_stats.to_csv(out_pattern_p, index=False)
df_stats
```

```python
rest_df_counts = rest_df.value_counts(subset="pattern")
rest_df_counts
```

```python
rest_df_stats = pd.DataFrame([rest_df_counts.index, rest_df_counts.tolist(), (rest_df_counts.tolist())]).T
rest_df_stats.rename(columns={0: "pattern", 1: "count", 2: "frequency"}, inplace=True)
rest_df_stats.frequency = rest_df_stats.frequency /  rest_df.shape[0]
rest_df_stats.to_csv(out_rest_pattern_p, index=False)
rest_df_stats
```
