# This notebook transforms the nutrients ingredient data into a matrix for clustering

### Connecting to the postgresql server

```python
import pandas as pd
import numpy as np
import sqlalchemy as sal

from sqlalchemy import text
```

#### Connect to the database

```python
nourish_user = ""

nourish_pswd = ""

engine = sal.create_engine('postgresql+psycopg2://' + nourish_user + ':' + nourish_pswd + '@awesome-hw.sdsc.edu/nourish')
conn = engine.connect()
```

#### Pull the list of ingredients by product

```python
query_ingredients = text('''SELECT "NDB_Number", array_agg("ingredient")
from "product_ingredients_2018"
group by "NDB_Number" ''')

result = conn.execute(query_ingredients)
ingredients_data = [i for i in result]
ingredients_data = pd.DataFrame(ingredients_data)
ingredients_data.set_index("NDB_Number", inplace = True)
ingredients_data.head(2)
```
